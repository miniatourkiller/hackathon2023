```python
arrzero = []
for i in range(totalrows):
    rows = filtered.iloc[i]
    arrzero.append(rows[0])
arrzero
```

```python
arr = []
arr2 = []
for i in range(totalrows):
    rows = filtered.iloc[i]
    arr2 = rows[0].split()
    for j in range(len(arr2)):
        if arr2[j].lower() != "ack" and arr2[j].lower() != "ace":
            arr.append(arr2[j].lower())
arr
```

```python
from spellchecker import SpellChecker
spell = SpellChecker()
spell.word_frequency.load_words(arrzero)
spell.word_frequency.load_words(arr)
```
